```python
from bidi.algorithm import get_display
from wordcloud import WordCloud

import matplotlib.pyplot as plt
import arabic_reshaper
import pandas as pd
import re
import string
```

```python
data = pd.read_excel("arquestionsaboutiraq.xlsx", index_col=0)
data.columns = ["question"]
data.head()
```

## Remove Diacritics

```python
arabic_diacritics = re.compile("""
                             ّ    | # Tashdid
                             َ    | # Fatha
                             ً    | # Tanwin Fath
                             ُ    | # Damma
                             ٌ    | # Tanwin Damm
                             ِ    | # Kasra
                             ٍ    | # Tanwin Kasr
                             ْ    | # Sukun
                             ـ     # Tatwil/Kashida
                         """, re.VERBOSE)

def remove_diacritics(text):
    text = re.sub(arabic_diacritics, '', text)
    return text
```

## Normalize

```python
def normalize_arabic(text):
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ؤ", "ء", text)
    text = re.sub("ئ", "ء", text)
    text = re.sub("ة", "ه", text)
    text = re.sub("گ", "ك", text)

    unwant_words = ["كان", "ماذا", "هو", "هل", "لماذا", "على", "كيف"]
    for i in unwant_words:
      text = re.sub(i, "", text)

    return text
```

## Remove Punctuations

```python
def remove_punctuations(text):
  arabic_punctuations = '''`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ'''
  english_punctuations = string.punctuation
  punctuations_list = arabic_punctuations + english_punctuations

  translator = str.maketrans('', '', punctuations_list)
  return text.translate(translator)
```

## Generate Word Cloud

```python
text = " ".join(data["question"].tolist())
text = normalize_arabic(text)
text = remove_punctuations(text)
text = remove_diacritics(text)
```

```python
def arabic_word_cloud(text):
  text = text
  wordcloud = WordCloud(font_path='arabic_2.otf', collocations=False).generate(text)
  
  return wordcloud.to_image()
```
